```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load IPL dataset directly from GitHub
deliveries = pd.read_csv("https://raw.githubusercontent.com/PriyaVerma06/EDA-IPL/fix-E1/data/deliveries.csv")
matches = pd.read_csv("https://raw.githubusercontent.com/PriyaVerma06/EDA-IPL/fix-E1/data/matches.csv")

print("Matches Shape:", matches.shape)
print("Deliveries Shape:", deliveries.shape)
```

```python
# Count number of wins by each team
team_wins = matches['winner'].value_counts()

print(team_wins)
```

```python
# Basic stats
total_matches = matches.shape[0]
total_teams = matches['winner'].nunique()
top_team = team_wins.idxmax()
top_wins = team_wins.max()

print("Total Matches:", total_matches)
print("Total Teams:", total_teams)
print("Top Team:", top_team)
print("Top Wins:", top_wins)
```

```python
#visualisation
plt.figure(figsize=(12,6))
team_wins.plot(kind='bar')

plt.title("Number of Matches Won by Each Team")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,8))
team_wins.head(5).plot(kind='pie', autopct='%1.1f%%')

plt.title("Top 5 Teams Win Share")
plt.ylabel("")

plt.show()
```

## Insight Summary

- Mumbai Indians and Chennai Super Kings are among the top-performing teams
- A few teams dominate the majority of wins, showing imbalance in performance
- Several teams have significantly lower win counts, indicating weaker consistency
- The distribution of wins is skewed toward top teams
