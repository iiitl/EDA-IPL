```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/matches.csv')
df.shape
```

```python
df.head()
```

```python
# Matches where toss winner is also match winner
df['toss_win_match_win'] = df['toss_winner'] == df['winner']
```

```python
# percentage of matches won by toss winners
toss_win_match_win_percentage = (df['toss_win_match_win'].mean())*100
toss_win_match_win_percentage
```

```python
# Grouping toss_win_match_win by toss decision

dp = df.groupby('toss_decision')['toss_win_match_win'].mean()*100
dp
```

```python
# Grouping toss_win_match_win by season and toss decision
f = df.groupby(['season', 'toss_decision'])['toss_win_match_win'].mean().unstack()*100
```

```python
# Bar graph contains each season and toss decision with their corresponding toss win percentage

f.plot(kind='bar', figsize=(12,6))
plt.title('Toss Win Percentage by Season and Decision')
plt.xlabel('Season')
plt.ylabel('Toss Win Percentage')
plt.show()
```
