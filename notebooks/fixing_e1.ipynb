```python
matchesPerSeason = matches['season'].value_counts().sort_index()
print(matchesPerSeason)
```

```python
plt.figure(figsize=(10,6))
plt.title("matches per season")
plt.xlabel("season")
plt.ylabel("no of matches")
plt.xticks(rotation=45)
sns.barplot(x=matchesPerSeason.index, y=matchesPerSeason.values,color="yellow")
plt.show()
```
