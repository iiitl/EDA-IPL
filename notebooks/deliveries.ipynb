```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/EDA-IPL/data/deliveries.csv")
df.head()
```

```python
balls_per_match = df.groupby("match_id").size()

max_balls = balls_per_match.max()
print("Max balls in a match:", max_balls)

balls_per_match.sort_values(ascending=False).head()
```

```python
dismissals = df["dismissal_kind"].value_counts()
dismissals
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=dismissals.index, y=dismissals.values)
plt.xticks(rotation=45)
plt.title("Distribution of Dismissal Types")
plt.ylabel("Count")
plt.xlabel("Dismissal Type")
plt.show()
```

```python
dismissals.plot.pie(autopct='%1.1f%%', figsize=(7,7))
plt.title("Dismissal Distribution")
plt.ylabel("")
plt.show()
```

## Findings

- The maximum number of balls in a match is higher than the standard 240 deliveries.
- This is due to extras such as wides and no-balls, which add additional deliveries.
- No major anomalies were found beyond expected variations.

- The most common type of dismissal is "caught".
- Other common dismissals include bowled, lbw, and run out.
- This indicates that fielding plays a major role in wickets taken.
