### 5. Visualization: Matches per Season


```python
plt.style.use('ggplot') 
matches_per_season = df['season'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
matches_per_season.plot(kind='bar', color='midnightblue', edgecolor='black')

plt.title('IPL Tournament History: Matches per Season', fontsize=15)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('season_plot.png')
plt.show()
```
