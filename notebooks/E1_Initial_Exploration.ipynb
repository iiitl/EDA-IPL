```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
df = pd.read_csv('/matches.csv')

# 2. Metadata
print(f"Dataset Shape: {df.shape}")
print("\nColumn Names:")
print(df.columns.tolist())

# 3. Unique Seasons
unique_seasons = df['season'].nunique()
print(f"\nTotal Unique Seasons: {unique_seasons}")

# 4. Missing Values
print("\nMissing Values per Column:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# 5. Visualization: Matches per Season
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=df, palette='viridis')
plt.title('Number of Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.xticks(rotation=45)
plt.show()
```
