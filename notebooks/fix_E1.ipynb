```python
# 1. Setup and Imports

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2. Load Data

df = pd.read_csv("https://raw.githubusercontent.com/PriyaVerma06/EDA-IPL/main/data/matches.csv")
```

```python
# 3. Metadata

print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns.tolist())
```

```python
# 4. Basic Summary

unique_seasons = df['season'].nunique()
print("\nNumber of unique seasons:", unique_seasons)
```

```python
# 5. Missing Values

missing_values = df.isnull().sum()
missing_cols = missing_values[missing_values > 0]

print("\nColumns with missing values:\n")
print(missing_cols)
```

```python
# 6. Visualization

matches_per_season = df['season'].value_counts().sort_index()

plt.figure()
matches_per_season.plot(kind='bar')
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Matches Played Per Season")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
