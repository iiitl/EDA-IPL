Load data: Read data/matches.csv.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('matches.csv')
```

```python
df
```

Metadata: Output the shape of the dataframe and list all column names.

```python
print("Shape of dataset:", df.shape)
print("\nColumns:", df.columns.tolist())
```

Basic Summaries: How many unique seasons are in the dataset?

```python
print("\nNumber of unique Seasons:" ,df['season'].nunique())
```

Missing Values: Identify which columns contain missing values (NaN).

```python
print("\nMissing values:\n", df.isnull().sum())
```

Visualization: Plot a bar chart showing the number of matches played per season to ensure the data loaded correctly and isn't skewed.

```python
matches_per_season = df['season'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,6))
matches_per_season.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```
