```python
import pandas as pd

df = pd.read_csv('/content/matches.csv')
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
print("Unique Seasons:", df['season'].nunique())
print(df['season'].unique())
```

```python
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
import matplotlib.pyplot as plt

matches_per_season = df.groupby('season').size()

plt.figure(figsize=(12, 6))
matches_per_season.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Number of Matches Played Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('matches_per_season.png')  # save for PR screenshot
plt.show()
```
