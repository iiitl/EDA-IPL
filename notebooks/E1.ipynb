```python
import pandas as pd



deliveries_df = pd.read_csv("C:\\Users\\SOUMYA\\EDA-IPL\\data\\deliveries.csv")
matches_df = pd.read_csv("C:\\Users\\SOUMYA\\EDA-IPL\\data\\matches.csv")

df = pd.merge(deliveries_df, matches_df[['id', 'season']], left_on='match_id', right_on='id', how='left')

display(df.head())
```

```python
print("Shape of dataset:", df.shape)

print("\nColumn Names:")
for col in df.columns:
    print(col)
```

```python
# 3. Basic Summary
unique_seasons = df['season'].nunique()
print("\nTotal Unique Seasons:", unique_seasons)
```

```python
# 4. Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Visualization: Plot a bar chart showing the number of matches played per season to ensure the data loaded correctly and isn't skewed.
```

```python
import matplotlib.pyplot as plt
matches_per_season = df['season'].value_counts().sort_index()

plt.figure()
matches_per_season.plot(kind='bar')

plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Matches Played")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
